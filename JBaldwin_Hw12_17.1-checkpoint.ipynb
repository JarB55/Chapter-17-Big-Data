```python
import sqlite3
```

```python
connection = sqlite3.connect('books.db')
```

```python
import pandas as pd
```

```python
# A. Select all authorsâ€™ last names from the authors table in descending order.
pd.read_sql('SELECT last FROM authors ORDER BY last DESC', connection)
```

```python
# B. Select all book titles from the titles table in ascending order.
pd.read_sql('SELECT title FROM titles ORDER BY title', connection)
```

```python
# C. Use an INNER JOIN to select all the books for a specific author. Include the title,
# copyright year and ISBN. Order the information alphabetically by title.

# all titles with the author ID = 1
pd.read_sql("""SELECT title, copyright, titles.[isbn]
    FROM titles
    INNER JOIN author_ISBN
    ON titles.isbn = author_ISBN.isbn
    WHERE id = 1
    ORDER BY title""", connection)
```

```python
# D. Insert a new author into the authors table.
cursor = connection.cursor()
cursor = cursor.execute("""INSERT INTO authors (first, last)
                            VALUES ('Jarron', 'Baldwin')""")

pd.read_sql('SELECT id, first, last FROM authors', connection)
```

```python
# E.Insert a new title for an author. Remember that the book must have an entry
# in the author_ISBN table and an entry in the titles table.

cursor.execute("""INSERT INTO titles (isbn, title, edition, copyright)
                            VALUES ('0000000001', 'How to add your book to a DB',
                            '1', '2022')""")
cursor.execute("""INSERT INTO author_ISBN (id, isbn)
                            VALUES ('6','0000000001')""")

pd.read_sql('SELECT * FROM titles', connection)
```
